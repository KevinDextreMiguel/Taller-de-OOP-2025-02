# Ejercicio 1

```
import pandas as pd
```

```
class Venta:
  def __init__(self):
    self.__dfVentas = self.generar()

  def generar(self):
    Meses = ["Enero","Febrero","Marzo","Abril"]
    Ventas = [4100,4200,4300,4400]
    Gastos = [3100,3200,3300,3400]

    dic_ventas = {"Meses":Meses,"Ventas":Ventas,"Gastos":Gastos}

    df_ventas = pd.DataFrame(dic_ventas)
    return df_ventas

  def mostrar(self):
    return self.__dfVentas


objeto_venta = Venta()
objeto_venta.mostrar()
```

# Ejercicio 2

```
class Venta:
  def __init__(self):
    self.__dfVentas = self.generar()

  def generar(self):
    Meses = ["Enero","Febrero","Marzo","Abril"]
    Ventas = self.generar_valores()
    Gastos = self.generar_valores()

    dic_ventas = {"meses":Meses,"ventas":Ventas,"gastos":Gastos}

    df_ventas = pd.DataFrame(dic_ventas)
    return df_ventas

  def generar_valores(self,cantidad=4):
    lista = []
    for i in range(cantidad):
      lista.append(np.random.randint(1000,5000))
    return lista

  def mostrar(self):
    return self.__dfVentas

  def generar_balance(self):
    self.__dfVentas["balance"] = self.__dfVentas["ventas"] - self.__dfVentas["gastos"]

#
objeto_venta = Venta()
objeto_venta.generar_balance()
objeto_venta.mostrar()
```

# Ejercicio 3

```
class Cotizacion:
  def __init__(self):
    self.__dfCotizacion = pd.read_excel("cotizacion.xlsx")

  def mostrar(self):
    print(self.__dfCotizacion)

  def estadisticas(self):
    df_estadisticas = self.__dfCotizacion.describe()
    resultado = df_estadisticas.loc[["mean","min","max"]]
    return resultado

objeto_cotizacion = Cotizacion()
objeto_cotizacion.mostrar()
print("\n\nresultados de los valores estadìsticos")
objeto_cotizacion.estadisticas()
#
```

```
# Eliminar del DataFrame los pasajeros con edad desconocida.
df_nuevo = df.dropna(subset=["Age"])
```

```
df_nuevo.reset_index(drop=True,inplace=True)
```

```
df_nuevo
```

```
# Mostrar la edad promedio o media de las mujeres que viajaban en cada clase.
filtro = df_nuevo['Sex'] == "female"
df_nuevo[filtro].groupby("Pclass")["Age"].mean()
```

```
# Añadir una nueva columna booleana para ver si el pasajero era menor de edad o no.
df_nuevo["menor_edad"] = df_nuevo["Age"] < 18
```

```
df_nuevo.head(8)
```

```
# Mostrar el porcentaje de menores y mayores de edad que sobrevivieron en cada clase.
df_nuevo.groupby(["Pclass","menor_edad"])["Survived"].value_counts(normalize=True) * 100
```
