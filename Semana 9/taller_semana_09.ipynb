```
class Encomienda:
    dicc_encomiendas = {}

    def __init__(self, codigo, remitente,destinatario,direccion_entrega,peso_kg,volumen_m3,costo_kg,costo_m3):
        self.__codigo = codigo
        self.__remitente = remitente
        self.__destinatario = destinatario
        self.__direccion_entrega = direccion_entrega
        self.__peso_kg = peso_kg
        self.__volumen_m3 = volumen_m3
        self.__costo_kg = costo_kg
        self.__costo_m3 = costo_m3

    def calcular_costo_peso(self):
        return self.__peso_kg * self.__costo_kg

    def calcular_costo_volumen(self):
        return self.__volumen_m3 * self.__costo_m3

    def calcular_valor_envio(self):
        return self.calcular_costo_peso() + self.calcular_costo_volumen()

    def imprimir_encomienda(self):
        print("Remitente:", self.__remitente)
        print("Destinatario:", self.__destinatario)
        print("Dirección de Entrega:", self.__direccion_entrega)
        print("Peso en Kg:", self.__peso_kg)
        print("Volumen en m3:", self.__volumen_m3)
        print("Costo por Kg:", self.__costo_kg)
        print("Costo por m3:", self.__costo_m3)
        print("Valor a enviar:", self.calcular_valor_envio())

    def agregar_encomienda(self,objeto):
        self.dicc_encomiendas[objeto.codigo] = objeto

    def imprimir_encomiendas(self):
        for elemento in self.dicc_encomiendas:
            print("="*20)
            print("Encomienda", elemento)
            self.dicc_encomiendas[elemento].imprimir_encomienda()


    @property
    def codigo(self):
        return self.__codigo
    @codigo.setter
    def codigo(self, codigo):
        self.__codigo = codigo
    @property
    def remitente(self):
        return self.__remitente
    @remitente.setter
    def remitente(self, remitente):
        self.__remitente = remitente
    @property
    def destinatario(self):
        return self.__destinatario
    @destinatario.setter
    def destinatario(self, destinatario):
        self.__destinatario = destinatario
    @property
    def direccion_entrega(self):
        return self.__direccion_entrega
    @direccion_entrega.setter
    def direccion_entrega(self, direccion_entrega):
        self.__direccion_entrega = direccion_entrega
    @property
    def peso_kg(self):
        return self.__peso_kg
    @peso_kg.setter
    def peso_kg(self, peso_kg):
        self.__peso_kg = peso_kg
    @property
    def volumen_m3(self):
        return self.__volumen_m3
    @volumen_m3.setter
    def volumen_m3(self, volumen_m3):
        self.__volumen_m3 = volumen_m3
```

```
obj1 = Encomienda(1,"remitente1","destinatario1","direccion_entrega1",400,500,5,3)
obj2 = Encomienda(2,"remitente2","destinatario2","direccion_entrega2",200,300,2,2)
obj3 = Encomienda(3,"remitente3","destinatario3","direccion_entrega3",300,400,4,4)

obj1.agregar_encomienda(obj1)
obj2.agregar_encomienda(obj2)
obj3.agregar_encomienda(obj3)
obj3.imprimir_encomiendas()
```

```
# 1,Surco,100,50000,40000 -> 1,Surco,100,60000,40000
# 1,Comas,100,50000,40000 -> 1,Surco,100,60000,40000
# 1,Surco,100,50000,40000
# 1,Los Olivos,100,50000,40000
# 1,Surco,100,50000,40000

# Todos los terrenos que son de Surco su valor comercial aumenta en un 20%
```

```
class Terreno:
    dicc_terrenos = {}
    def __init__(self, codigo, ubicacion, area, valor_m2_comercial, valor_m2_predial):
        self.__codigo = codigo
        self.__ubicacion = ubicacion
        self.__area = area
        self.__valor_m2_comercial = valor_m2_comercial
        self.__valor_m2_predial = valor_m2_predial

    def calcular_valor_comercial(self):
        return self.__area * self.__valor_m2_comercial

    def calcular_valor_predial(self):
        return self.__area * self.__valor_m2_predial

    def calcular_margen_ganancia(self):
        return self.calcular_valor_comercial() - self.calcular_valor_predial()

    def imprimir_terreno(self):
        print("Ubicación:", self.__ubicacion)
        print("Área:", self.__area)
        print("Valor m2 Comercial:", self.__valor_m2_comercial)
        print("Valor m2 Predial:", self.__valor_m2_predial)
        print("Valor Comercial:", self.calcular_valor_comercial())
        print("Valor Predial:", self.calcular_valor_predial())
        print("Margen de Ganancia:", self.calcular_margen_ganancia())

    def agregar_terreno(self,objeto):
        self.dicc_terrenos[objeto.codigo] = objeto

    def imprimir_terrenos(self):
        for elemento in self.dicc_terrenos:
            print("="*20)
            print("Terreno", elemento)
            self.dicc_terrenos[elemento].imprimir_terreno()

    # Todos los terrenos que son de Surco su valor comercial aumenta en un 20%
    def modificar_valor_comercial(self):
        for elemento in self.dicc_terrenos:
            # Utilizamos el getter
            if self.dicc_terrenos[elemento].ubicacion == "Surco":
              # Utilizamos el getter

                # self.dicc_terrenos[elemento].get_valor_m2_comercial()
                aumento_con20_porciento = self.dicc_terrenos[elemento].valor_m2_comercial * 1.2

                # Utilizamos el setter
                self.dicc_terrenos[elemento].valor_m2_comercial = aumento_con20_porciento
                # self.dicc_terrenos[elemento].set_valor_m2_comercial(aumento_con20_porciento)

    @property
    def codigo(self):
        return self.__codigo
    @codigo.setter
    def codigo(self, codigo):
        self.__codigo = codigo

    @property
    def ubicacion(self):
        return self.__ubicacion
    @ubicacion.setter
    def ubicacion(self, ubicacion):
        self.__ubicacion = ubicacion

    @property
    def area(self):
        return self.__area
    @area.setter
    def area(self, area):
        self.__area = area

    @property
    def valor_m2_comercial(self):
        return self.__valor_m2_comercial
    @valor_m2_comercial.setter
    def valor_m2_comercial(self, valor_m2_comercial):
        self.__valor_m2_comercial = valor_m2_comercial

    @property
    def valor_m2_predial(self):
        return self.__valor_m2_predial
    @valor_m2_predial.setter
    def valor_m2_predial(self, valor_m2_predial):
        self.__valor_m2_predial = valor_m2_predial
```

```
obj1 = Terreno(1,"Surco",100,50000,4000)
obj2 = Terreno(2,"Comas",100,50000,4000)
obj3 = Terreno(3,"Surco",100,50000,4000)
obj4 = Terreno(4,"Los Olivos",100,50000,4000)
obj5 = Terreno(5,"Surco",100,50000,4000)

obj1.agregar_terreno(obj1)
obj2.agregar_terreno(obj2)
obj3.agregar_terreno(obj3)
obj4.agregar_terreno(obj4)
obj5.agregar_terreno(obj5)
obj5.imprimir_terrenos()
print("Despuès de incremnetor los precios en un distrito")
obj5.modificar_valor_comercial()
obj5.imprimir_terrenos()
```

```
class Pasajero:
  def __init__(self,numero_pasaporte, nombre, edad, tipo_asiento):
    self.__numero_pasaporte = numero_pasaporte
    self.__nombre = nombre
    self.__edad = edad
    self.__tipo_asiento = tipo_asiento

  def imprimir_pasajero(self):
    print("Numero de Pasaporte:", self.__numero_pasaporte)
    print("Nombre:", self.__nombre)
    print("Edad:", self.__edad)
    print("Tipo de Asiento:", self.__tipo_asiento)

  @property
  def numero_pasaporte(self):
    return self.__numero_pasaporte
  @numero_pasaporte.setter
  def numero_pasaporte(self, numero_pasaporte):
    self.__numero_pasaporte = numero_pasaporte

  @property
  def nombre(self):
    return self.__nombre
  @nombre.setter
  def nombre(self, nombre):
    self.__nombre = nombre

  @property
  def edad(self):
    return self.__edad
  @edad.setter
  def edad(self, edad):
    self.__edad = edad

  @property
  def tipo_asiento(self):
    return self.__tipo_asiento
  @tipo_asiento.setter
  def tipo_asiento(self, tipo_asiento):
    self.__tipo_asiento = tipo_asiento
```

```
class Vuelo:
  def __init__(self, numero_vuelo, hora_salida,hora_llegada, ciudad_partida,ciudad_llegada):
    self.__numero_vuelo = numero_vuelo
    self.__hora_salida = hora_salida
    self.__hora_llegada = hora_llegada
    self.__ciudad_partida = ciudad_partida
    self.__ciudad_llegada = ciudad_llegada

    self.__pasajeros_primera_clase = []
    self.__pasajeros_turistas= []

  def agregar_pasajero(self,pasajero):
    if pasajero.tipo_asiento == "Primera Clase":
      if len(self.__pasajeros_primera_clase) < 8:
        self.__pasajeros_primera_clase.append(pasajero)
      else:
        print("No se puede agregar más pasajeros a la primera clase")
    else:
      if len(self.__pasajeros_turistas) < 92:
        self.__pasajeros_turistas.append(pasajero)
      else:
        print("No se puede agregar más pasajeros a la clase Turista")

  def imprimir_vuelo(self):
    print("Numero de Vuelo:", self.__numero_vuelo)
    print("Hora de Salida:", self.__hora_salida)
    print("Hora de Llegada:", self.__hora_llegada)
    print("Ciudad de Partida:", self.__ciudad_partida)
    print("Ciudad de Llegada:", self.__ciudad_llegada)

  def imprimir_pasajeros_primera_clase(self):
    print("Pasajeros de Primera Clase:")
    for pasajero in self.__pasajeros_primera_clase:
      pasajero.imprimir_pasajero()

  def imprimir_pasajeros_turistas(self):
    print("="*20)
    print("Pasajeros de Turista:")
    for pasajero in self.__pasajeros_turistas:
      pasajero.imprimir_pasajero()

  @property
  def numero_vuelo(self):
    return self.__numero_vuelo
  @numero_vuelo.setter
  def numero_vuelo(self, numero_vuelo):
    self.__numero_vuelo = numero_vuelo

  @property
  def hora_salida(self):
    return self.__hora_salida
  @hora_salida.setter
  def hora_salida(self, hora_salida):
    self.__hora_salida = hora_salida
```

```
pasajero1 = Pasajero("123456789", "Juan", 30, "Primera Clase")
pasajero2 = Pasajero("987654321", "Maria", 25, "Turista")
pasajero3 = Pasajero("456789123", "Pedro", 40, "Primera Clase")
pasajero4 = Pasajero("321654987", "Ana", 20, "Turista")

vuelo1 = Vuelo("V001", "12:00", "14:00", "Lima", "Buenos Aires")
vuelo1.agregar_pasajero(pasajero1)
vuelo1.agregar_pasajero(pasajero2)
vuelo1.agregar_pasajero(pasajero3)
vuelo1.agregar_pasajero(pasajero4)

vuelo1.imprimir_vuelo()
vuelo1.imprimir_pasajeros_primera_clase()
vuelo1.imprimir_pasajeros_turistas()
```
