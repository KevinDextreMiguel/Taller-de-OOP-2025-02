```
# Clases Abstractas
from abc import ABC, abstractmethod

class Figura(ABC): # Base (Padre)
    @abstractmethod
    def area(self):
        pass

    def mostrar_mensaje(self):
      print("Esto es una figura")

class Cuadrado(Figura):# Hija (derivada) (subclase)
    def __init__(self, lado):
        self.__lado = lado

    def area(self):
        return self.__lado ** 2

    def __str__(self):
        return f"Cuadrado de lado {self.__lado}"

    def mostrar_mensaje(self):
      print("Esto es un cuadrado")


objeto_cuadrado = Cuadrado(5)
print(objeto_cuadrado)
print(objeto_cuadrado.area())
objeto_cuadrado.mostrar_mensaje()
#
```

```
class Persona: #base
  def __init__(self, nombre, edad):
    self.nombre = nombre
    self.edad = edad

  def mostrar_informacion(self):#derivada
    print(f"Nombre: {self.nombre}, Edad: {self.edad}")

class Empleado(Persona):
  def __init__(self, nombre, edad, salario):
    super().__init__(nombre, edad)
    self.salario = salario

  def mostrar_informacion(self):
    print(f"Nombre: {self.nombre}, Edad: {self.edad}")
    print(f"Salario: {self.salario}")
```

```
objeto_persona = Persona("Juan", 25)
objeto_persona.mostrar_informacion()
```

```
objeto_empleado = Empleado("Juan", 25, 5000)
objeto_empleado.mostrar_informacion()
#
```

```
from abc import ABC, abstractmethod

class Accionar:
  def encender(self):
    print("Encendido")
  def apagar(self):
    print("Apagado")
```

```
class Electrodomestico: #Base
  def __init__(self, potenciaKW, marca):
    self.potenciaKW = potenciaKW
    self.marca = marca

  @abstractmethod
  def ConsumoEnergetico(self):
    pass

  def __str__(self):
    return f"Potencia: {self.potenciaKW}, Marca: {self.marca}"

  def encender(self,objeto_accionar):
    if isinstance(objeto_accionar,Accionar):
        objeto_accionar.encender()
    else:
        print("El objeto no es de tipo Accionar")

  def apagar(self,objeto_accionar):
    if isinstance(objeto_accionar,Accionar):
        objeto_accionar.apagar()
    else:
        print("El objeto no es de tipo Accionar")
```

```
class Coccion(Electrodomestico):
  def __init__(self,potenciaKW, marca,programable):
    super().__init__(potenciaKW, marca)
    self.programable = programable

  def ConsumoEnergetico(self,costo_kw=10):
    return self.potenciaKW * costo_kw

  def __str__(self):
    return f"Potencia: {self.potenciaKW}, Marca: {self.marca}, Programable: {self.programable}"
```

```
class Cocina(Coccion):
  def __init__(self,potenciaKW, marca,programable,hornillas):
    super().__init__(potenciaKW, marca,programable)
    self.hornillas = hornillas

  def ConsumoEnergetico(self,costo_kw=10,costo_hornilla=5):
    return super().ConsumoEnergetico(costo_kw) + self.hornillas * costo_hornilla

  def __str__(self):
    return f"Potencia: {self.potenciaKW}, Marca: {self.marca}, Programable: {self.programable}, Hornillas: {self.hornillas}"
```

```
class Horno(Coccion):
  def __init__(self,potenciaKW, marca,programable,capacidad):
    super().__init__(potenciaKW, marca,programable)
    self.capacidad = capacidad

  def ConsumoEnergetico(self,costo_kw=20,costo_capacidad=10):
    return super().ConsumoEnergetico(costo_kw) + self.capacidad * costo_capacidad

  def __str__(self):
    return f"Potencia: {self.potenciaKW}, Marca: {self.marca}, Programable: {self.programable}, capacidad: {self.capacidad}"
```

```
# Crear objetos
objeto_accionar = Accionar()

# objeto de tipo cocina
objeto_cocina = Cocina(20, "Oster",True,6)
print(objeto_cocina)
print(objeto_cocina.ConsumoEnergetico())
objeto_cocina.encender(objeto_accionar)
objeto_cocina.apagar(objeto_accionar)
```

```
# Crear objetos
objeto_accionar = Accionar()

# objeto de tipo horno
objeto_horno = Horno(40,"Phillip",False,10)
print(objeto_horno)
print(objeto_horno.ConsumoEnergetico())
objeto_horno.encender(objeto_accionar)
objeto_horno.apagar(objeto_accionar)
```

```
from abc import ABC, abstractmethod
import pandas as pd

class Volumen(ABC):
  def __init__(self,id, name):
    self.__id = id
    self.__name = name

  @property
  def id(self):
    return self.__id
  @id.setter
  def id(self,id):
    self.__id = id # Fixed: Assign to the private attribute

  @property
  def name(self):
    return self.__name

  @abstractmethod
  def __str__(self):
    pass
```

```
class Book(Volumen):
  def __init__(self,id, name, isbn_code):
    super().__init__(id, name)
    self.__isbn_code = isbn_code

  def __str__(self):
    return f"ID: {self.id}, Name: {self.name}, ISBN: {self.__isbn_code}"
```

```
class Newpaper(Volumen):
  def __init__(self,id, name, issn_code):
    super().__init__(id, name)
    self.__issn_code = issn_code

  def __str__(self):
    return f"ID: {self.id}, Name: {self.name}, ISSN: {self.__issn_code}"
```

```
import pandas as pd

class Library:
  def __init__(self,name,address):
    self.name = name
    self.address = address
    self.volumns = []

  def add_volumen(self,volumen):
    if self.existe_volumen(volumen):
      print("El volumen ya existe")
    else:
      self.volumns.append(volumen)

  def existe_volumen(self,volumen):
    for v in self.volumns:
      if v.id == volumen.id:
        return True
    return False

  def remove_volumen(self,id):
    for volumen in self.volumns:
      if volumen.id == id:
        #self.volumns.pop(id)
        self.volumns.remove(volumen)

  def show_volumns(self):
    for volumen in self.volumns:
      print(volumen)

  def guardar_excel(self):
    data_to_save = []
    for volumen in self.volumns:
      data_to_save.append({"ID": volumen.id, "Name": volumen.name})

    df = pd.DataFrame(data_to_save)
    df.to_excel("ejrcicio.xlsx", index=False)
```

```
# Crear objetos de libro
libro1 = Book(1, "Libro 1", "1234567890")
libro2 = Book(2, "Libro 2", "0987654321")
libro3 = Book(3, "Libro 3", "5624456666")

#Crear objetos peri√≤dico
periodico1 = Newpaper(4, "Periodico 1", "1234567890")
periodico2 = Newpaper(5, "Periodico 2", "0987654321")
periodico3 = Newpaper(6, "Periodico 3", "5624456666")

#Craer un objeto de volumen
biblioteca = Library("Biblioteca", "Calle 123")

# Agregar libros a la biblioteca
biblioteca.add_volumen(libro1)
biblioteca.add_volumen(libro2)
biblioteca.add_volumen(libro3)

biblioteca.add_volumen(periodico1)
biblioteca.add_volumen(periodico2)
biblioteca.add_volumen(periodico3)

biblioteca.show_volumns()
```
